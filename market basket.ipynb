```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
data = pd.read_csv('D:\datasets\gro.csv')
item = data['Item(s)']
# item = item[:40]
len(data)
```

```python
d1 = data.drop(['Item(s)'], axis = 1)
# d1 = d1.loc[:39]

len(d1)
```

```python
records = []
for i in range(0,len(d1)):
        records.append([str(d1.values[i,j]) for j in range(0,32)]) 
#print(records)
len(records)
```

```python
a = records
b = item
c = []

for i in range(0,len(item)):
    for j in range(0,b[i]):
        c.append(a[i])
```

```python
len(c)
```

```python
import datetime
```

```python
x = datetime.datetime.now()
asrule = apriori(c, min_support=0.0045, min_confidence=0.2, min_lift=1.5, min_length=3)
y = datetime.datetime.now()
asres = list(asrule)
```

```python
y-x
```

```python
antecedent = []
consequent = []
support = []
confidence = []
lift = []
p = 1
```

```python
for item in asres:                                        

    # first index of the inner list
    # Contains base item and add item
    print(p)
    p = p+1
    pair = item[0] 
    items = [x for x in pair]
     
    print("Rule: " ,items[0] , " -> " , items[1])
    antecedent.append(str(items[0]))
    consequent.append(str(items[1]))
    #second index of the inner list
    print("Support: " + str(item[1]))
    support.append(str(item[1]))
    #third index of the list located at 0th
    #of the third index of the inner list
    confidence.append(str(item[2][0][2]))
    lift.append(str(item[2][0][3]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
  
```

```python
final_res = pd.DataFrame({'antecedent':antecedent,'conesequent':consequent,'support':support,'confidence':confidence,'lift':lift})
```

```python
final_res.sort_values(by=['confidence'],ascending=True)
```
